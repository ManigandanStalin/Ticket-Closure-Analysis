```python
##import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
df=pd.read_csv('E:\Data Set\incident_event_log.csv')
```

```python
df1=pd.DataFrame(df)
```

```python
closed=df1[df1['incident_state']=='Closed']  #We need only data on closed tickets for our analysis. 
closed.head()
```

```python
closed.drop(labels=['incident_state','active','sys_created_by','sys_created_at','sys_updated_by','sys_updated_at','contact_type','cmdb_ci','impact','urgency','knowledge','u_priority_confirmation','notify','rfc','vendor','caused_by','problem_id'], axis=1, inplace =True) 
#Feature selecting(Remove unwanted columns)
```

```python
closed.head(5)
```

```python
## replace special character with Null
for i in closed:
    closed[i]=closed[i].replace('?',np.nan)
    print(closed[i])
```

```python
closed.isna().sum()
```

```python
## Replace Null values with mode in all variable
for column in closed:
    closed[column].fillna(closed[column].mode()[0], inplace=True)
```
